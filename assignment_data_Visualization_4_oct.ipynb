# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python


import numpy as np
import pandas as pd

np.random.seed(42)

n_samples = 100
feature1 = np.random.normal(0, 1, n_samples)
feature2 = feature1 * 0.5 + np.random.normal(0, 0.5, n_samples)  # Correlated with feature1
feature3 = feature1 * 0.3 + feature2 * 0.7 + np.random.normal(0, 0.5, n_samples)  # Correlated with feature1 and feature2

data = pd.DataFrame({
    'Feature 1': feature1,
    'Feature 2': feature2,
    'Feature 3': feature3
})
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 10,  # Bubble size proportional to population
            alpha=0.5,
            edgecolors='w',
            linewidth=2)

# Add annotations for each country
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 fontsize=10,
                 ha='center')

# Set labels and title
plt.title('Bubble Chart of Countries: GDP vs Population')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.grid(True)

# Show plot
plt.show()
```

# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Output to notebook (if using a Jupyter Notebook)
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y values as the sine of x
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave', x_axis_label='X', y_axis_label='Sine of X', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

# Show the plot
show(p)
```

# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Output to notebook (if using a Jupyter Notebook)
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure
p = figure(x_range=fruits, title='Fruit Counts', x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)

# Add vertical bars to the figure
p.vbar(x='fruits', top='counts', source=source, width=0.9, color='navy')

# Add labels to the bars
for fruit, count in zip(fruits, counts):
    p.text(x=fruit, y=count + 1, text=[str(count)], text_align='center', text_font_size='10pt')

# Show the plot
show(p)
```

# 4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Output to notebook (if using a Jupyter Notebook)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title='Histogram of Normally Distributed Data',
           x_axis_label='Value',
           y_axis_label='Count',
           width=800,
           height=400)

# Add bars to the histogram using quad
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color='navy', line_color='white', alpha=0.7)

# Show the plot
show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)

y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LogColorMapper


output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

heatmap_data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'values': data_heatmap.flatten()
}

p = figure(title='Heatmap Example',
           x_axis_label='X',
           y_axis_label='Y',
           width=600,
           height=600)

mapper = linear_cmap(field_name='values', palette='Viridis256', low=min(heatmap_data['values']), high=max(heatmap_data['values']))

p.rect(x='x', y='y', width=0.1, height=0.1, source=heatmap_data,
        line_color=None, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
